# stress-controlled : pressure vs volumetric strain 

```python
lj2GPa=160.2 # convert lj reduced units to GPa

load_lj_cac = np.arange(0,12,1)
load_GPa_cac =load_lj_cac*lj2GPa # load is in reduced units

patom_cac = read('data.initial',format='cac-data',style="cac")
patom_cac.set_calculator(cac_calc)
fixstrain = np.ones((3,3))
pstress = patom_cac.get_cell()*0.0


for i in range(0,len(load_GPa_cac)):
    pstress[0][0] = load_GPa_cac[i]
    pstress[1][1] = load_GPa_cac[i]
    print('stress=',load_GPa_cac[i])
    pbox = stressbox(patom_cac, express=pstress, fixstrain = fixstrain, atom_style='cac',Grain_natom=9)
    dyn=FIRE(pbox, dt=0.01, maxmove=0.01, dtmax=0.1)
#     dyn=MDMin(pbox, dt=0.01)
    dyn.run(fmax=0.001,steps=10000)
    write(str(i)+".sq.cac_data",patom_cac, format='cac-data',atom_style="cac")
    write(str(i)+".sq.cac_lammpsdata",patom_cac, format='lammps-data',atom_style='atomic')



vol_sq_cac = np.zeros(len(load_lj_cac))
pe_sq_cac = np.zeros(len(load_lj_cac))

for i in range(0,len(load_GPa_cac)):
    sq_atom_cac = read(str(i)+'.sq.cac_data',format='cac-data',style="cac")
    sq_atom_cac.set_calculator(cac_calc)
    vol_sq_cac [i] = sq_atom_cac.cell.volume
    pe_sq_cac[i] = sq_atom_cac.get_potential_energy()


fig,ax = plt.subplots(figsize=(7,5)) 
ax.set_xlabel('Volumetric strain')
ax.set_ylabel('Pressure')

ax.plot((vol_sq_cac[0]-vol_sq_cac)/vol_sq_cac[0],load_lj_cac,'g^--',label='square CAC')

ax.legend()
fig.show()

```
